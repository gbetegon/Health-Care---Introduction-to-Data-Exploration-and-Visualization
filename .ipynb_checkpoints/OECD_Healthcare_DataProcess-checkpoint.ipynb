## 2. Functions <a id='2.'></a>

```python
def df_structure(df):
    """returns the lenght and unique values in each column of a given dataframe"""
    print "Dataset contains {} rows and has the following structure:\n".format(len(df))
    for col in df.ix[:, df.columns != 'Value']:
        print col, df[col].unique(), "\n"

def count_ctries():
    """Counts the number of unique countries in a dataframe"""
    dfs = [quality, cost, inflation]
    countries = []
    for df in dfs:
        country = df['LOCATION'].unique().tolist()
        countries.append(country)
    return countries
    
def print_ctries_in_df():
    """Prints results of function count_ctries""" 
    print ("quality, cost and inflation dfs have "
           "{}, {} and {} countries respectively.\n"
           .format(len(countries[0]),
                   len(countries[1]),
                   len(countries[2]))
          )
    
def uncommon_ctries():
    """returns countries that are not common to two dataframes"""
    non_common = (list(set(countries[0]) - set(countries[1]) - set(countries[2])) +
                  list(set(countries[2]) - set(countries[1]) - set(countries[0]))
                 )
    return non_common

def plot_missing_data(df):
    """plots count of missing variables entries for each year in a dataframe,
    being 35 the maximum ammount on variables that can be possibly missing on a year"""
    var_recount = df.groupby(level=2, axis=1).count()
    for column in var_recount.columns:
        var_recount[column] = var_recount[column].map(lambda x: 35 - x)
    var_recount.plot(kind='bar', width = 1.2, figsize=(18,2.5))
    plt.legend(loc='upper center', ncol=4)
    plt.title('Missing Data by Year and Variable', fontsize=12, fontweight='bold')
    plt.ylabel('Number of variables missing')
    plt.xlabel('Years')
    plt.ylim([0,20])
    plt.show()
```

## 3. Data Process <a id='3.'></a>

### 3.A Import data <a id='3.1'></a>

Our data is stored in different csv files downloaded from the OECD and World Bank websites. The first step is to import the data into Pandas.

We will import the following three sources to three dataframes (quality, cost, and inflation):
* Health Quality
   * Variables: Life Expectancy at Birth, Mortality, Infant Mortality, Maternal Mortality,
       * Source: [OECD/Healthcare Quality Indicators](http://www.oecd-ilibrary.org/social-issues-migration-health/data/oecd-health-statistics/oecd-health-data-health-care-quality-indicators_data-00592-en?isPartOf=/content/datacollection/health-data-en)
* Health Cost
   * Variables: Healthcare cost per capita, healthcare gdp share
       * Source: [OECD/Health Expenditure and Financing](http://www.oecd-ilibrary.org/social-issues-migration-health/data/oecd-health-data-health-expenditure-and-financing_health_e_f-data-en)
   * Variables: Inflation
       * Source: [World Bank/World Development Indicators](http://databank.worldbank.org/data/reports.aspx?source=2&series=NY.GDP.DEFL.KD.ZG&country=)

```python
# Import csv files to pandas
directory = '/Users/Stephanie/desktop/thinkful/U5-capstone/data/'

docs = ('OECD_health_quality.csv',
        'OECD_health_expenditure.csv',
        'OECD_inflation.csv')

dfs = []

for doc in docs:
    df = pd.read_csv(directory + doc)
    dfs.append(df)

quality, cost, inflation = [dfs[i] for i in range(0,3)]
```

### 3.B Reduce country selection to OECD members <a id='3.2'></a>

The scope of our study is the 35 OECD economies. We're going to drop any other countries included in our just created dataframes. Before we do that we will standardize the column name that contains the country codes in each dataframe.

```python
# rename quality df 'COU' column
quality.rename(columns={'COU':'LOCATION'}, inplace=True)

# Count number of countries in each df
countries = count_ctries()
print_ctries_in_df()
```

Cost df is already limited to the 35 OECD members. We only have to find the countries are not common to  the three dataframes and drop them.

```python
# List countries
non_common = uncommon_ctries()
print ("The following countries aren't common and need to be removed: {}"
       .format(', '.join(non_common))
      )
            
# Remove uncommon countries
quality = quality[~quality['LOCATION'].isin(non_common)]
cost = cost[~cost['LOCATION'].isin(non_common)]
inflation = inflation[~inflation['LOCATION'].isin(non_common)]

# Recount again number of countries
countries = count_ctries()
print_ctries_in_df()
```

### 3.C Prepare Dataframes <a id='3.3.'></a>
We just dropped some countries that we didn't need for the study. We're now going to do the same thing with the data contained for each country: we're going to look at the structure of each dataframe and get rid of anything we don't need to create something more manageable.

#### 3.C.a Prepare Cost dataframe <a id='3.3.1'></a>

```python
df_structure(cost)
```

There are more than 400k rows of information and tens of different long-named variables. We will select only the variables we need and rename them to lighten our code

```python
# slice down rows and columns
fin_scheme = 'All financing schemes'
function = 'Current expenditure on health (all functions)'
provider = 'All providers'
measures = ['Share of gross domestic product', 'Per capita, current prices']
renamed = ['GDP share', 'Cost pc']

cost = cost[(cost['Financing scheme'] == fin_scheme)&
            (cost['Function'] == function) &
            (cost['Provider'] == provider)
           ]

cost = cost[cost['Measure'].isin(measures)]
cost = cost.replace(measures, renamed)
cost = cost[['Year', 'LOCATION', 'Measure', 'Value']]
cost.rename(columns={'Measure':'Variable'}, inplace=True)

print "Dataframe has now {} rows. This is how it looks".format(len(cost))
cost.head(3)
```

We are going to add the inflation to the cost dataframe. By comparing healthcare cost's yearly increases to the inflation we can determine if the former is becoming more or less expensive in comparison to the rest of goods and services. If the differential between health care cost and inflation is consistently positive throughout a time series, we can conclude that the former is becoming more expensive relatively to the rest of goods and services.

We want to make sure when concatenating inflation df to cost df that we don't miss data on the way. So we'll check that we have the same number of rows before and after this operation.

```python
# add inflation to cost df
pd.options.mode.chained_assignment = None #disables warning
inflation = inflation[['TIME', 'LOCATION', 'MEASURE', 'Value']]
inflation.columns = cost.columns
inflation['Variable'] = 'Inflation'

# check no lost rows
print "total rows in both dataframes separately = {}".format((len(inflation) + len(cost)))
cost = pd.concat([cost, inflation], ignore_index='True')
print "total rows in consolidated dataframe = {}".format(len(cost))
```

We're done with the cost df for the moment. This is how the head of the df looks at this point

```python
cost['Category'] = "Health Expenses"
cost.head(3)
```

```python
df_structure(quality)
```

```python
# slice down rows and columns
variables = ['Total population at birth',
             'All causes of death',
             'Infant mortality, No minimum threshold of gestation period or birthweight',
             'Maternal mortality'
            ]

measures = ['Years',
            'Deaths per 100 000 population (standardised rates)',
            'Deaths per 1 000 live births',
            'Deaths per 100 000 live births']

quality = quality[quality['Variable'].isin(variables)]
quality = quality.replace(variables,
                          ['Life exp',
                           'Mort',
                           'Inf mort',
                           'Mat mort']
                         )

quality = quality[quality['Measure'].isin(measures)]
quality = quality[['Year', 'LOCATION', 'Variable', 'Value']]
quality['Category'] = "Health Quality"
print "Dataframe has now {} rows. This is how it looks".format(len(quality))
quality.head(3)
```

#### 3.H.b Fill remaining NaN values using moving averages <a id='3.8.2'></a>

Let's plot again the missing data after substituting the maternal mortality data.

```python
plot_missing_data(data)
```

The peak on missing values is still concentrated in the 2014 Mortality rate, but we also have information missing for other years/variables. To solve this, we're going to use moving averages.

Since the length of a gap might vary (from one missing value to several concatenated missing values), we're going to try to use the smallest possible window for every single gap.

To do that we will loop through the dataframe starting with a smaller window moving into a bigger one to fill the gaps that remain after each loop.

```python
for window in range(1,5):
    values = pd.DataFrame.rolling(data, min_periods=1,window=window,center=False).mean().shift(0)
    data = data.fillna(values)

for window, shift in zip(range(1,5), reversed(range(-2, 1))):
    values = pd.DataFrame.rolling(data, min_periods=1,window=window,center=False).mean().shift(shift)
    data = data.fillna(values)
```

After this operation we can see how plotting the missing data won't return any results

```python
plot_missing_data(data)
```

### 3.I Calculate yearly cost increase <a id='3.9'></a>
We are going to calculate the yearly cost increase in healthcare cost using the healthcare cost in each country's national currency.


We'll save the results as a new column of the dataframe. After including the new column we need to sort the dataframe multi-indexed columns. Otherwise, we will have problems when slicing the dataframe in the future.

```python
for country in data.columns.levels[0]:
    cost_increase = (country, 'Health Expenses', 'Cost increase')
    price = (country,'Health Expenses','Cost pc')
    data[cost_increase] = (data[price].pct_change(periods=1))*100

#sort dataframe
data = data.sortlevel(axis=1)
levels = data.columns.lexsort_depth
print 'cost increase included in df'
print 'dataframe sorted on {} levels'.format(levels)
```

We don't have the previous year to calculate the increment in 2000. We'll just use an average from the next two years.

```python
values = pd.DataFrame.rolling(data, min_periods=1,window=2,center=False).mean().shift(-2)
data = data.fillna(values)
```

### 3.J Country grouping by GDP income <a id='3.10'></a>
The GDP income is the main indicator of country economic development. There are very significant differences based on GDP pc between countries in our study population and we want to see how that affects the results.

We'll create three initial groups: low, high and medium GDP income pc (per capita). The distribution will be the following: low (less than $22.500 pc), medium (between $22.500 and $45.000) and high gdp pc (above $45.000).

```python
# Plot GDP pc distribution
ctries = data.columns.levels[0]

GDP = pd.read_csv(directory + 'WB_GDP_pc_constant.csv', header=2, index_col='Country Code')
GDP = GDP.loc[ctries, '2000':'2014']
avg_GDP = GDP.mean(axis=1)

plt.figure()
avg_GDP.sort_values(inplace=True)
colors = ["red" if avg < 22500 else "yellow" if avg < 45000 else "green" for avg in avg_GDP]
avg_GDP.plot(kind='bar', color=colors)
plt.axhline(y=22500, xmin=0, xmax=1, c='r')
plt.axhline(y=45000, xmin=0, xmax=1, c='yellow')
plt.title('GDP per capita distribution', fontsize=12, fontweight='bold')
plt.ylabel('Dollars')
plt.xlabel('Countries')
plt.show()
```

We want to include these groups in our dataframe aggregating and averaging all countries in each group. The US data won't be used for the grouping

```python
# Drop out US from list and group by GDP pc 
avg_noUS = avg_GDP[avg_GDP != avg_GDP['USA']]

gdp_groups = {
              'low': avg_noUS[avg_noUS <22500].index.tolist(),
              'medium': avg_noUS[(avg_noUS >= 22500) & (avg_noUS < 45000)].index.tolist(),
              'high': avg_noUS[avg_noUS >= 45000].index.tolist()
             }

categories = data.columns.get_level_values(1)[0:9]
variables = data.columns.get_level_values(2)[0:9]

idx = pd.IndexSlice
for group, values in gdp_groups.iteritems():
    print "Average values for {} gdp group have been included".format(group)
    for cat, var in zip(categories, variables):
        data[group, cat, var] = data.loc[:, idx[values, cat, var]].mean(axis=1)
```

The US is in the high GDP pc braket. Within the group there's a clear gap between the four countries with the highest GDP per capita and the rest of the group.

Since the study focuses on the US economy, we're also going to add a new group, composed by the 8 US neighbor countries, attending to their GDP income. That is, those with the closest GDP pc values to the US.

```python
# Group by US GDP neighbors
US_gdp_neighbors = avg_GDP[(26-4):(26+5)].index.tolist()
US_gdp_neighbors.remove('USA')

for cat, var in zip(categories, variables):
    data['US GDP Neighbors', cat, var] = data.loc[:, idx[US_gdp_neighbors, cat, var]].mean(axis=1)
print "Average values for US GDP Neighbors have been included".format(group)
```

### 3.K Calculate differential between health care cost increase and inflation <a id='3.11'></a>

### 3.L Obtain Health Quality Composite Indicator <a id='3.12'></a>

We have collected data for a series of indicators that we have grouped under the 'Health Quality' category. These indicators are: life expectancy, mortality, maternal mortality and infant mortality.

Although these indicators don't only depend on how good is a country's healthcare, improving its results should be the goal of every healthcare system. This is, increasing average life expectancy and lowering mortality rates. For this reason they can be good indicators of a country's healthcare quality. 

But at the moment we just have these four separate variables. We need to create an indicator that sums up the information of the four variables and provides a score that can help us create rank the results. To achieve this we need to go over the following steps:

- Transform Health Quality Indicators into a logarithmic scale
- Normalize the outcome for each variable between a 0-1 scale
- Weight and merge the values into a final Health Quality Score

#### 3.L.a Transform Health Quality variables to logarithmic scale <a id='3.12.1'></a>

As said, we want to transform our variables into a logarithmic scale. The transformation compresses the values on the extremes of our range, reducing it, but without having a huge impact on the rest of our data. This way we can keep the detail we're is more important and avoid really high values to distort the relation between different results.

```python
# Log transform Health Quality
qlty = data.loc[:, (slice(None), 'Health Quality')]
qlty = np.log(qlty.astype('float64'))
```

To see the effect that the logarithmic transformation produces, we're going to plot the before and after. We can see how 

In the figure below there's a comparison of each country/variable without and with a log transform. As we can see, the log transform helps smooth extreme values that could other wise distort our scores.

For example we have two countries for maternal mortality whose values during the first years are more than 8 eight times higher than the average. Without a log transform, if we were to convert this variable to a 0-1 scale, the vast majority of our population would be compressed in values very close to 0, and we would lose a lot of detail.

```python
# Compare variables before/after log transform
qlty_vars = ['Life exp', 'Mat mort', 'Inf mort', 'Mort']
ctries = qlty.columns.levels[0].tolist()

sns.set_style('darkgrid')

plt.figure()
fig, ax = plt.subplots(1, 4, figsize=(14,2.2))
for var, ax in zip(qlty_vars, ax):
        ax.plot(data.loc[:, (slice(None), 'Health Quality', var)], marker='.', markersize=5, linewidth=0.1)
        ax.set_title(var)
        ax.grid(b=None, axis='x')
plt.show()

plt.figure()
fig, ax = plt.subplots(1, 4, figsize=(14,2.2))
for var, ax in zip(qlty_vars, ax):
        ax.plot(qlty.loc[:, (slice(None), 'Health Quality', var)], marker='.', markersize=5, linewidth=0.1)
        ax.set_title('Log ' + var)
        ax.grid(b=None, axis='x')
        if var == 'Life exp':
            ax.legend(ctries, bbox_to_anchor=(2.4, -0.5), loc=8, borderaxespad=0., ncol=13, fontsize='x-small')
plt.show()
```

#### 3.L.b Normalize variables <a id='3.12.2'></a>

Our variables use different scales. We want to normalize the data so all variables use the same scale and can be compared and aggregated.

```python
for var in qlty_vars:
    Min = qlty.loc[:, (slice(None), slice(None), var)].values.min()
    Max = qlty.loc[:, (slice(None), slice(None), var)].values.max()
    x = qlty.loc[:, (slice(None), slice(None), var)]
    if var == 'Life exp':
        qlty.loc[:, (slice(None), slice(None), var)] = (x - Min) / (Max - Min)
    else:
        qlty.loc[:, (slice(None), slice(None), var)] = (x - Max) / (Min - Max)
                     
print qlty['USA']
```

#### 3.L.c Weight and aggregate into composite quality score <a id='3.12.3'></a>
All our variables are now measured on a 0-1 scale. Now we need to assign a weight to each variable and aggregate them. Life expectancy will make for 50% of the indicator; maternal mortality for 25%; and the other two variables will explain 12.5% of the final score.

The maximum score will be a 100 points

```python
ctries = qlty.columns.levels[0]
scores = pd.DataFrame(index=qlty.index)

for ctry in ctries:
    life_exp = qlty[ctry, 'Health Quality', 'Life exp']
    mort = qlty[ctry, 'Health Quality', 'Mort']
    mat_mort = qlty[ctry, 'Health Quality', 'Mat mort']
    inf_mort = qlty[ctry, 'Health Quality', 'Inf mort']
    scores[ctry] = (0.5*life_exp + 0.25*mort + 0.125*mat_mort + 0.125*inf_mort) * 100
```

We now have a score for each country. For example, these are the results for the US

```python
scores['USA']
```
